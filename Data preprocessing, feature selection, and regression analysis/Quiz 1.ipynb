3. Codifique todos los atributos categóricos, seleccione el método de codificación y explique porqué  selecciona cada método.



Un método para para los datos categoricos nominales es el target enconding. Esta codificación consiste en reemplazar cada etiqueta con el promedio de la variable objetivo para esa etiqueta. También se puede utilizar el método One-Hot, pero se estaría trabajando con una cantidad de columnas nuevas que se crean muy grande por lo se dificulta el manejo de los datos.

Se utilizará la libreria category_encoders de sci-kit learn para realizar la codificación de variables categoricas. Especificament, TargetEncoder para realizar la codificación de las variables

```python

from sklearn.impute import SimpleImputer
import category_encoders as ce


# Crear DataFrame con la columna 'price'
target = pd.DataFrame(df['price'])

# Reemplazar valores '?' por NaN
target['price'] = target['price'].replace('?', np.nan)

# Crear instancia del imputador y especificar la estrategia de imputación
imputer = SimpleImputer(strategy='mean')

# Aplicar imputación a la columna 'price'
target['price'] = imputer.fit_transform(target[['price']])

nominal_columns = ['make', 'fuel-type', 'aspiration', 'body-style', 'engine-location', 'drive-wheels', 'engine-type', 'fuel-system']

encoder = ce.TargetEncoder(cols=nominal_columns)
df = encoder.fit_transform(df, target['price'])

#Imprimir database modificada
df.head()

```

Ahora, para los atributos ordinales se realiza un reemplazo de los valores. Es decir, en el caso de num-of-doors se reemplazaron los valores "four" por el valor numérico 4 y "two" por el valor numérico 2. Esto se hizo para convertir las categorías de la cantidad de puertas de los vehículos en valores numéricos que representen la cantidad real, lo igual aplica para num-of-cylinders.

```python
# Mapeo de valores en la columna num-of-doors
df['num-of-doors'] = df['num-of-doors'].replace({'four': 4, 'two': 2})

# Mapeo de valores en la columna num-of-cylinders
df['num-of-cylinders'] = df['num-of-cylinders'].replace({'four': 4, 'six': 6, 'five': 5, 'eight': 8, 'two': 2, 'three': 3, 'twelve': 12})

#Imprimir database modificada
df.head()
```

4. Realice un análisis de datos faltantes en todo el conjunto de datos y aplique imputación para  habilitar esos registros. (8 puntos)


Para impuntar los datos faltantes dentro de las columnas, se utiliza SimpleImputer para que reemplace esos valores por el valor de la media de la columna. Esto debido a que con la media se preserva la distribución y minimiza el impacto en las estadísticas

```python

# Crear instancia de la estrategia de imputación como 'mean'
imputer = SimpleImputer(strategy='mean')

# Obtener las columnas con valores faltantes
columnas_con_faltantes = df.columns[df.isnull().any()].tolist()

# Imputar los valores faltantes en cada columna
for columna in columnas_con_faltantes:
    df[columna] = imputer.fit_transform(df[[columna]])

# Verificar si hay valores faltantes después de la imputación
print(df.isnull().sum())

df.head()
```

5. Escale los datos (5 puntos)

Algunos de los algoritmos comunes de escalado de datos incluyen la estandarización y la normalización.

La normalización, es un proceso en el cual los datos son transformados para que estén en un rango específico, como por ejemplo entre 0 y 1. Esto es útil cuando se quiere mantener la escala relativa de los datos y se desea evitar que variables con valores más altos dominen sobre aquellas con valores más bajos.

La estandarización, por otro lado, es un proceso en el cual los datos son transformados de tal manera que tengan una media de cero y una desviación estándar de uno. Esto es útil cuando se trabaja con algoritmos que asumen una distribución normal de los datos o cuando se desea comparar diferentes variables que tienen diferentes unidades de medida. Sabiendo que los datos contienen outliers, este metodo es util para eliminar el impacto de los mismos.
